# Mixing Factorの個数を求める
- N,Kをかえれば他の場合でも計算できる

```julia
#SC用の凍結ビット集合 符号長:N=32,情報ビット数:K=21
include("../FrozenBitsSelect.jl")
N = 32; K= 16
F = frozenbits_select_FN(N,K)

#=
Mixing Factorの個数を求める
計算方法：
    maxFを求める、凍結ビットの最大値
    maxFより小さい情報ビットのインデックスの数を求める
        1:maxFで凍結ビットの集合に含まれていない数
=#
maxF = maximum(F)
mc = filter(x->!in(x,F),1:maxF)
println("maxF:$(maxF)")
println("($N,$K)のMixing Factorの個数は$(length(mc))")
println(mc)
```

# 生成行列の重み分布をもとめる

```julia
#生成行列の重みの分布を把握する
include("../PolarEncode.jl")
N = 32
n = Int(log2(N))
G = kron_generate_G(n)
w = zeros(Int,N)
for i=1:N
    w[sum(G[i,:])] += 1
end

#K=21は無理そう, K=17で考える
println(w)
using Plots
plot(w,line=:stem,yticks=[1:10;],xticks = [1,2,4,8,16,32],label="",marker=:circle)
```
